# Netflix Titles – Exploratory Analysis & Insights

```

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('netflix_titles.csv')
df.head()
```

## Data Cleaning

```

df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')

df['release_decade'] = (df['release_year']//10)*10
df['is_movie'] = df['type'].apply(lambda x: 1 if x=='Movie' else 0)
```

## Visualizations

```

# Pie chart: Movie vs TV Show
df['type'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()

# Content released per year
df.groupby('release_year').size().plot()
plt.show()

# Top 10 genres
df['listed_in'].str.split(', ').explode().value_counts().head(10).plot(kind='bar')
plt.show()

# Country vs content heatmap
country_content = df['country'].value_counts().head(10)
sns.heatmap(country_content.to_frame(), annot=True, cmap='coolwarm')
plt.show()
```
